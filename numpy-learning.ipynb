### Array Attributes
- ndim 
- shape
- size 
- itemsize
- dtype

```python
a1 = np.arange(10,dtype=np.int32)
a2 = np.arange(12,dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#ndim
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
#shape
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
#size
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
#itemsize
print(a1.itemsize) #prints 4 bcz there is the datatype int32 which takes 4 bytes for each item
print(a2.itemsize)
print(a3.itemsize)
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

### Array Operation
- Array with scalar
    - Arithmetic (+ , - , * , / , **)
    - Relational (<= , >= , < , > , ==)
- Array with vectors (+ , - , * , / , **)

```python
a1 = np.arange(20).reshape(4,5)
a2 = np.arange(20,40).reshape(4,5)
print(a1)
print(a2)
```

```python
# Array with scalar
# Arithmetic

print(a1+2)
print(a1-2)
print(a1*2)
print(a1/2)
print(a1**2)
```

```python
# Relational

print(a1<= 5)
print(a1>=5)
print(a1==7)
print(a1<10)
```

```python
# Arrays operation with vectors

print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)
print(a1**a2)
```

```python
a1 = np.round(np.random.random((3,3))*100)
a1
```

```python
print(np.sum(a1))
print(np.max(a1))
print(np.min(a1))
print(np.prod(a1))
print(np.max(a1,axis = 0)) # gives the max elem from each column
print(np.max(a1,axis = 1)) # gives the max elem from each row
```
