### Array Operation
- Array with scalar
    - Arithmetic (+ , - , * , / , **)
    - Relational (<= , >= , < , > , ==)
- Array with vectors (+ , - , * , / , **)

```python
a1 = np.arange(20).reshape(4,5)
a2 = np.arange(20,40).reshape(4,5)
print(a1)
print(a2)
```

```python
# Array with scalar
# Arithmetic

print(a1+2)
print(a1-2)
print(a1*2)
print(a1/2)
print(a1**2)
```

```python
# Relational

print(a1<= 5)
print(a1>=5)
print(a1==7)
print(a1<10)
```

```python
# Arrays operation with vectors

print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)
print(a1**a2)
```

### Array Functions
- max / min / sum / prod
- mean / median / std / var
- sin / cos / tan
- np.dot
- log / exp
- round / floor / ceil

```python
a1 = np.round(np.random.random((3,3))*100)
a1
```

```python
# max / min / sum / prod

print(np.sum(a1))
print(np.max(a1))
print(np.min(a1))
print(np.prod(a1))
print(np.max(a1,axis = 0)) # gives the max elem from each column
print(np.max(a1,axis = 1)) # gives the max elem from each row
```

```python
# mean / median / std / var

print(np.mean(a1))
print(np.var(a1))
print(np.std(a1))
print(np.median(a1))
print(np.max(a1,axis = 0)) # gives the mean of each column
print(np.max(a1,axis = 1)) # gives the mean of each row
```

```python
# sin / cos / tan

np.sin(a1)
```

```python
#dot product
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
dot = np.dot(a1.T,a2)
print(dot)
```

```python
# log / exp

print(np.log(a1))
print(np.exp(a1))
```

```python
# floor/ceil/round

print(np.ceil(a1))
print(np.floor(a1))
print(np.round(np.random.random((4,5))*100))
```

### Indexing and Slicing

- Indexing is a[0,2,1...]
- slicing is a[0:2,:,:]

```python
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
# extract 2nd 2d array using slicing
a3[1,:,:]
```

```python
#extract [[0,2],[18,20]]

a3[::2,0,::2]
```

```python
# extract [[22, 23],[25, 26]]

a3[2,1:,1:]
```

### Stacking
- hstack
- vstack

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
print(a1)
print()
print(a2)
```

```python
# hstack (horizontal stacking)
np.hstack((a1,a2))
```

```python
#vstack (vertical stacking)
np.vstack((a1,a2))
```
